# Задача о планировании тренировок 
# (в первом приближении)

Модель представляет из себя расписание тренировок для команды.

#### __Мотивация__: 

в студенческих командах зачастую ограничены ресурсы, а именно:
- имеется ограниченное количество тренировок (слишком малое)
- игрокам не позволяет расписание всегда присутствовать на всех тренировках

Существующие тренировочные программы рассчитаны на больший ресурс, поэтому в описанном положении становятся менее эффективными

#### __Решение__: 

составить такое расписание, которое будет учитывать расписание игроков и их индивидуальные потребности в развитии по
трем (возможно их больше) направлениям: 
- техника 
- физическая подготовка 
- командные взаимодействия

Таким образом, расписание должно расставлять упражнения в том порядке, в котором они принесут наибольшую пользу игрокам.

В качестве *параметров* модели обсуждаются вместе с игроками и тренером их потребности в различных упражнениях в процентном соотношении, а также собственно то, насколько различные упражнения закрывают ту или иную потребность

В качестве *переменных* определяется собственно расписание упражнений и нижняя граница удовлетворенности участниками процесса

В качестве *целевой функции* будет та же самая нижняя граница.

#### Параметры:
- L - Количество слотов в одной тренировке (размер слота - 5 минут)
- D - Количество тренировочных дней
- P - Количество тренировок в неделю
- W = D ÷ P - количество тренировочных недель
- I - Множество игроков
- J - Множество упражнений
- C - Множество классов упражнений
- Comp - множество дней соревнований
- t - Длительность упражнений (t[j] - количество слотов, занимаемых упражнением j)
- alpha - принадлежность упражнения j к классу c (упражнение может быть полезным по разному)
- gamma - потребность игрока i в упражнениях класса c
- a - способность игрока присутствовать на тренировке

#### Переменные:
- x_dj, d = 1:D, j = 1:J - на день d ставится упражнение j - xdj = 1, иначе - xdj = 0
- S - нижняя граница удовлетворенности игроков

Ограничения:
- sum_j=1:J(x_dj*t_j) <= L, d = 1:D - в каждый момент времени идет только одно упражнение
- S_i >= S, i = 1:I - минимальная удовлетворенность не ниже порога

#### Целевая функция:
- S -> max
